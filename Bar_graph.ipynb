```
import matplotlib.pyplot as plt
import numpy as np
x=([1,2,3,4,5,6])
y=([300,250,450,360,700,300])
plt.title("Beautiful gragh",fontname="Candara",fontsize=30,color="grey")
plt.xlabel("Time\n(Years)",fontsize=20)
plt.ylabel("Earnings\n(kes)",fontsize=20)
plt.bar(x,y)
plt.show()

```
